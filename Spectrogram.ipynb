```python
import IPython.display as ipd
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import librosa 
import pandas as pd

gun_shot_file= "C:/Users/User/OneDrive/Masaüstü/Voices/fold1/7061-6-0-0.wav"
dog_bark_file="C:/Users/User/OneDrive/Masaüstü/Voices/fold1/7383-3-0-0.wav"
dog_bark2_file="C:/Users/User/OneDrive/Masaüstü/Voices/fold1/7383-3-0-1.wav"
dog_bark3_file="C:/Users/User/OneDrive/Masaüstü/Voices/fold1/7383-3-1-0.wav"

ipd.Audio(gun_shot_file)
#ipd.Audio(dog_bark_file)
#ipd.Audio(dog_bark2_file)
#ipd.Audio(dog_bark3_file)

#load audio files with librosa
gun_shot , sr = librosa.load(gun_shot_file)
dog_bark , _ = librosa.load(dog_bark_file)
dog_bark2 ,_= librosa.load (dog_bark2_file)
dog_bark3,_ =librosa.load(dog_bark3_file)

#extracting short-time fourier transform
FRAME_SIZE=2048
HOP_SIZE=512
S_gun_shot = librosa.stft( gun_shot ,n_fft=FRAME_SIZE , hop_length=HOP_SIZE)
S_gun_shot.shape
type(S_gun_shot[0][0])
# Calculating the spectrogram
Y_gun_shot= np.abs(S_gun_shot)**2
S_gun_shot.shape
type(Y_gun_shot[0][01  ])

#visualizing the spectrogram

def plot_spectrogram(Y , sr , hop_length , y_axis= 'linear'):
    plt.figure(figsize=(25,10))
    librosa.display.specshow(Y,
                             sr=sr,
                             hop_length=hop_length,
                             x_axis='time',
                             y_axis=y_axis)
    plt.colorbar(format='%+2.f')
plot_spectrogram(Y_gun_shot, sr, HOP_SIZE)
# log-amplitude spectrogram
Y_log_gun_shot= librosa.power_to_db(Y_gun_shot)
plot_spectrogram(Y_log_gun_shot, sr, HOP_SIZE)

#log-frequency spectrogram
plot_spectrogram(Y_log_gun_shot, sr, HOP_SIZE, y_axis="log")

# other voice's spectrograms
S_dog_bark = librosa.stft( dog_bark,n_fft=FRAME_SIZE , hop_length=HOP_SIZE)
S_dog_bark2 = librosa.stft(  dog_bark2 ,n_fft=FRAME_SIZE , hop_length=HOP_SIZE)
S_dog_bark3 = librosa.stft( dog_bark3 ,n_fft=FRAME_SIZE , hop_length=HOP_SIZE)

Y_dog_bark= librosa.power_to_db(np.abs(S_dog_bark))**2
Y_dog_bark2= librosa.power_to_db(np.abs(S_dog_bark2))**2
Y_dog_bark3= librosa.power_to_db(np.abs(S_dog_bark3))**2

plot_spectrogram(Y_dog_bark, sr, HOP_SIZE, y_axis="log")
plot_spectrogram(Y_dog_bark2, sr, HOP_SIZE, y_axis="log")
plot_spectrogram(Y_dog_bark3, sr, HOP_SIZE, y_axis="log")







```
